## Answers

1. The command `python -m venv venv` creates a virtual environment named venv in the current directory. This isolates dependencies from the global Python installation.

    1.1
    - `pip list` - lists all installed Python packages in the current environment, along with their versions.
    - `pip freeze > requirements.txt` - saves a list of installed Python packages and their versions to a file named requirements.txt, for example it can be used for dependencies sharing.
    - `pip install -r requirements.txt` - installs all the packages listed in requirements.txt.

2. Work with commands
- `conda env list` - lists all available Conda environments on the system.
- `conda create -n env_name python=3.5` - creates a new Conda environment named env_name with Python 3.5.
- `conda env update -n env_name -f file.yml` - updates the Conda environment env_name using the dependencies specified in file.yml.
- `source activate env_name` - activates the Conda environment env_name. (conda activate env_name)
- `source deactivate` - deactivates the currently active Conda environment. (conda deactivate)
- `conda clean -a` - cleans up unused packages.

6. Import dependencies and env configuration

- `pip install -r requirements.txt` - installs packages listed in requirements.txt using pip.
- `conda env create -f environment.yml` - creates a new Conda environment from environment.yml.

7. Work with packages

- `pip list` -  displays a list of installed pip packages.
- `pip show <package>`- shows detailed information about a specific pip package.
- `conda list` - lists all installed packages in the current Conda environment.
