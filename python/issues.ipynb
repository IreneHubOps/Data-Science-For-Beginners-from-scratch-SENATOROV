Список вопросов по работе с Issues на GitHub


Общие вопросы

1. Что такое Issues на GitHub и для чего они используются? - вид документа используемый в GitHub  для общения по каки-либо проблемам, вопросам, обучение, пояснение работы кода
2. Чем Issues отличаются от других инструментов управления задачами?
Issues можно связывать с коммитами, пул-реквестами и другими компонентами репозитория.
3. Какие основные компоненты (поля) есть у каждого Issue?
     * - Название: Краткое описание задачи или проблемы.
     * - Описание: Подробное объяснение проблемы или задачи.
     * - Labels: Метки для классификации и фильтрации.
     * - Assignees: Ответственные за выполнение задачи.
     * - Comments: Обсуждения, связанные с Issue.
     * - Milestone: Этап или цель, к которой относится Issue.

Создание Issues

4. Как создать новое Issue в репозитории?
    *   выделить код, copy permalink, new issues
    *   Issues -> New Issue -> Submit new issue.

5. Какие данные рекомендуется указывать в описании Issue для лучшего понимания задачи? 
 что надо сделать, файл, трассировка ошибки, скрин, ссылка на видео с ошибкой, ожидаемый результат

6. Какие теги (labels) можно добавить к Issue? Какие из них стандартные? ug, enhancement, question, documentation
7. Как прикрепить Assignees (ответственных) к Issue? справа вверху Assignees выбрать


Работа с Issues

8. Как использовать Labels для классификации задач? применять их при создании issues. Для поска Issues -> labels
9. Для чего нужен Milestone, и как связать его с Issue? Milestone - это дедлайн выполнения. Привязывается справа при создании issues
10. Как привязать Issue к пул-реквесту (Pull Request)? Внести в pull request ссылку на issues после #
11. Как добавить комментарий к существующему Issue?
    Откройте Issue и прокрутите вниз до раздела комментариев, где можно ввести текст и отправить его.


Закрытие и завершение Issues

12. Как закрыть Issue вручную?-зайти в файл в гитхаб,карандаш, внести измемения.Commit changes - привязать комит к проблеме. Указать имя issue (#number) Closes #number issue->Sign off and commit changes
13. Можно ли автоматически закрыть Issue с помощью сообщения в коммите или пул-реквесте? Как это сделать?
    да. commit body or description(issue link) Closes issue link

14. Как повторно открыть закрытое Issue, если работа ещё не завершена?
    Issue -> Reopen issue


Фильтрация и поиск

15. Как найти все открытые или закрытые Issues в репозитории?
На странице Issues можно выбрать "Open" или "Closed"
16. Как использовать фильтры для поиска Issues по меткам, исполнителям или другим критериям?
Использовать фильтры в верхней части страницы Issues, чтобы выбрать метки, авторов, состояния

Интеграции и автоматизация

18. Как настроить автоматические уведомления о новых или изменённых Issues?
В настройках репозитория подписаться на уведомления о событиях Issues
19. Что такое Projects в контексте GitHub, и как связать их с Issues?
Projects — это инструменты для управления проектами. Issues можно связывать с карточками в Projects для удобного отслеживания их статуса.
20. Какие сторонние инструменты можно использовать для автоматизации работы с Issues (например, боты, Webhooks)? GitHub Actions для автоматизации рабочих процессов, или интеграции с другими сервисами (например, JIRA, Trello).


Коллаборация

21. Как упомянуть другого пользователя в комментарии к Issue?

22. Как запросить дополнительные данные или уточнения у автора Issue?

23. Что делать, если Issue неактуально или его нужно объединить с другим?
Если Issue неактуально, можно его закрыть. Если его нужно объединить с другим, можно оставить комментарий в обоих Issue и закрыть один из них с ссылкой на другой.


Практические аспекты

24. Как использовать шаблоны для создания Issues?
добавить файл ISSUE_TEMPLATE в директорию .github.
25. Что такое Linked Issues, и как создать связь между задачами?
это возможность связывать разные задачи друг с другом. Это можно сделать, добавив комментарий с упоминанием другого Issue (например, "related to #123").
26. Какие метрики (например, время выполнения) можно отслеживать с помощью Issues?
время выполнения, общее количество открытых и закрытых Issues, скорость выполнения задач
27. Какие best practices рекомендуются при работе с Issues в команде?
создавать четкие и подробные описания, использовать метки для организации, назначать ответственных
