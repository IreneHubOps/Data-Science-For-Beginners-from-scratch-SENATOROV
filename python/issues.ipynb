#### Общие вопросы
1. Что такое Issues на GitHub и для чего они используются?  
    Issues на GitHub — это инструмент для отслеживания задач, ошибок, улучшений и других аспектов проекта. Они позволяют пользователям сообщать о проблемах, предлагать новые функции, обсуждать изменения и координировать работу над проектом. Issues могут быть связаны с конкретными ветками кода, пользователями и метками, что помогает организовать и структурировать работу.
2. Чем Issues отличаются от других инструментов управления задачами?    
    Issues на GitHub интегрированы непосредственно в репозиторий и могут быть связаны с конкретными коммитами, ветками и пул-реквестами. Это позволяет более тесно связать задачи с кодом и отслеживать их выполнение на каждом этапе разработки. В отличие от отдельных инструментов управления задачами, такие как Jira или Trello, Issues предоставляют удобный способ для разработчиков и пользователей работать с задачами прямо в контексте кода.
3. Какие основные компоненты (поля) есть у каждого Issue?   
    Каждый Issue в GitHub имеет несколько основных компонентов, которые помогают эффективно управлять задачами и проектами. Вот основные из них:
    1. **Title (Заголовок)**: Краткое описание задачи или проблемы.
    2. **Description (Описание)**: Подробное описание задачи, включая контекст, шаги для воспроизведения (если это баг), ожидаемый результат и любые дополнительные сведения.
    3. **Labels (Метки)**: Теги, которые помогают классифицировать и фильтровать задачи. Примеры: bug, feature, enhancement, help wanted, good first issue и т.д.
    4. **Assignees (Ответственные)**: Участники, назначенные для выполнения задачи. Это может быть один или несколько человек.
    5. **Milestone (Этап)**: Этап или версия, к которой относится задача. Это помогает отслеживать прогресс по проекту.
    6. **Linked Pull Requests (Связанные пул-реквесты)**: Пул-реквесты, связанные с задачей. Это помогает отслеживать, какие изменения были внесены для решения задачи.
    7. **Comments (Комментарии)**: Обсуждение задачи, вопросы, ответы и любые дополнительные сведения, которые могут быть полезны для понимания задачи.
#### Создание Issues
4. Как создать новое Issue в репозитории?   
    - Перейдите на страницу репозитория.
    - Нажмите на кнопку "Issues" в верхней части страницы.
    - Нажмите на кнопку "New issue" (Новая задача).
    - Заполните заголовок и описание задачи.
    - Добавьте метки (labels), если необходимо.
    - Укажите этап (milestone), если задача связана с определенной версией или этапом проекта.
    - Нажмите на кнопку "Submit new issue" (Отправить новую задачу).
5. Какие данные рекомендуется указывать в описании Issue для лучшего понимания задачи?  
    - **Заголовок**: Краткое и информативное описание задачи.
    - **Описание**: Подробное описание задачи, включая контекст, шаги для воспроизведения (если это баг), ожидаемый результат и фактический результат.
6. Какие теги (labels) можно добавить к Issue? Какие из них стандартные?    
    - **Теги (labels)**: Метки, которые помогают классифицировать задачи. Стандартные метки могут включать:
        - `bug` (ошибка)
        - `enhancement` (улучшение)
        - `feature` (новая функция)
        - `documentation` (документация)
7. Как прикрепить Assignees (ответственных) к Issue?    
    - В разделе "Assignees" (Ответственные) можно указать одного или нескольких пользователей, которые будут ответственны за выполнение задачи.
    - Нажмите на поле "Assignees" и выберите пользователей из списка.
    - Если вы хотите назначить задачу себе, выберите свой никнейм из списка.
#### Работа с Issues
8. Как использовать Labels для классификации задач? 
    Используйте метки для классификации задач по категориям, таким как тип задачи (баг, улучшение, документация), приоритет, статус и т.д.
9. Для чего нужен Milestone, и как связать его с Issue? 
    - **Milestone** (этап) — это цель или версия, к которой привязаны задачи. Он помогает организовать работу по конкретным этапам или версиям продукта.
    - Чтобы связать Issue с Milestone, перейдите в раздел "Milestone" и выберите нужный этап из списка.
    - Это поможет отслеживать прогресс по конкретным этапам и планировать релизы.
10. Как привязать Issue к пул-реквесту (Pull Request)?  
    - В разделе "Linked pull requests" выберите нужный пул-реквест из списка.
    - Это поможет связать задачу с конкретной реализацией и отслеживать прогресс по ней.
11. Как добавить комментарий к существующему Issue? 
    - Перейдите в раздел "Comments" под Issue.
    - Напишите ваш комментарий в текстовом поле и нажмите "Comment" или "Submit comment".
    - Комментарии помогают обсуждать задачу, делиться информацией и координировать работу.
#### Закрытие и завершение Issues
12. Как закрыть Issue вручную?  
    - Перейдите в раздел "Close issue" и выберите причину закрытия из списка.
13. Можно ли автоматически закрыть Issue с помощью сообщения в коммите или пул-реквесте? Как это сделать?   
    - Да, можно использовать ключевые слова в комментариях к коммитам или пул-реквестам, такие как "close #123" или "fixes #123".
    - Это автоматически закроет указанный Issue после слияния пул-реквеста.
14. Как повторно открыть закрытое Issue, если работа ещё не завершена?  
    - Перейдите в раздел "Reopen issue" и нажмите "Reopen".

    - Используйте фильтры в разделе "Issues" и выберите "Open" или "Closed".
16. Как использовать фильтры для поиска Issues по меткам, исполнителям или другим критериям?    
    - Используйте фильтры в разделе "Issues" и выберите нужные параметры.
    - Например, можно выбрать метки, исполнителей или другие критерии для фильтрации Issues.
17. Как сортировать Issues по приоритету, дате создания или другим параметрам?  
    - Используйте фильтры в разделе "Issues" и выберите нужные параметры для сортировки.
    - Например, можно выбрать "Sort by" и выбрать нужный параметр для сортировки Issues.
#### Интеграции и автоматизация
18. Как настроить автоматические уведомления о новых или изменённых Issues? 
    - Используйте настройки уведомлений в разделе "Notifications" и настройте уведомления по нужным параметрам.
    - Например, можно настроить уведомления о новых Issues, изменениях в Issues или комментариях к Issues.

20. Какие сторонние инструменты можно использовать для автоматизации работы с Issues (например, боты, Webhooks)?

#### Коллаборация
21. Как упомянуть другого пользователя в комментарии к Issue?   
    - В комментариях к Issue можно упомянуть другого пользователя, добавив символ "@" перед его именем. Например, "@username". Это отправит уведомление пользователю о вашем комментарии.
    - Также можно использовать упоминания в описании Issue, чтобы привлечь внимание нужных людей.
22. Как запросить дополнительные данные или уточнения у автора Issue?   
    - Если вам нужны дополнительные данные или уточнения от автора Issue, вы можете оставить комментарий с соответствующим запросом. Например, "Пожалуйста, уточните следующее: ..." или "Мне нужны дополнительные детали по поводу ...".
    - Вы также можете использовать метки или ассоциировать Issue с другими задачами, чтобы привлечь внимание нужных людей.
23. Что делать, если Issue неактуально или его нужно объединить с другим?   
    - Если Issue неактуально, вы можете закрыть его, добавив соответствующий комментарий. Например, "Закрываю, так как это неактуально." или "Закрываю, так как это дубликат #123."
    - Если Issue нужно объединить с другим, вы можете закрыть его как дубликат, указав номер другого Issue. Например, "Закрываю как дубликат #123."
#### Практические аспекты
24. Как использовать шаблоны для создания Issues?   
    В корне репозитория создать папку .github. В ней создать папку ISSUE_TEMPLATE. Внутри папки ISSUE_TEMPLATE создайте файл для каждого шаблона, который вы хотите использовать. Файлы шаблонов должны иметь расширение .md (Markdown). 

    Например: 
    - bug_report.md - шаблон для багов 
    - feature_request.md - шаблон для запросов на добавление функционала 

    При создании Issue выбираем один из доступных шаблонов, которые мы создали, заполняем поля шаблона и создаем issue.
