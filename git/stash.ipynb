1. Что делает команда git stash?

Ответ:
git stash — сохраняет изменения в локальном репозитории, но не отправляет их на удалённый репозиторий.

2. Как просмотреть список всех сохранённых изменений (стэшей)?

Ответ:
git stash list

3. Какая команда применяется для использования верхнего стэша?

Ответ:
git stash apply

4. Как применить конкретный стэш по его номеру?

Ответ:
git stash apply stash@{NUMBER}

5. Чем отличается команда git stash apply от git stash pop?

Ответ:
git stash apply - применяет изменения из стэша, но не удаляет его.
git stash pop - применяет и удаляет стэш.

6. Что делает команда git stash drop?

Ответ:
Удаляет указанный стэш.

7. Как полностью очистить все сохранённые стэши?

Ответ:
git stash clear

8. В каких случаях удобно использовать git stash?

Ответ:
Когда нужно сделать merge, переключиться на другую ветку без коммита или временно спрятать изменения.

9. Что произойдёт, если выполнить git stash pop, но в проекте есть конфликтующие изменения?

Ответ:
Git выдаст конфликт, и его нужно будет решить вручную.

10. Можно ли восстановить удалённый стэш после выполнения git stash drop?

Ответ:
Да, с помощью git fsck --unreachable, если стэш не перезаписан.

11. Что делает команда git stash save "NAME_STASH"?

Ответ:
Сохраняет изменения в именованный стэш.

12. Что делает команда git stash apply stash@{NUMBER}?

Ответ:
Применяет указанный стэш без удаления.

13. Что делает команда git stash pop stash@{NUMBER}?

Ответ:
Применяет указанный стэш и удаляет его.
