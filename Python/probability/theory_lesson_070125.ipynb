```python
# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É —Å —Ç—Ä–µ–º—è –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∞–º–∏
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))

# 1. –ì—Ä–∞—Ñ–∏–∫ E(œâ) ‚Üí R —Å Œµ‚ÇÅ, Œµ‚ÇÇ, Œµ‚ÇÉ, Œµ‚ÇÑ
ax1.plot([0, 1, 2, 3], [0.5, 0.7, 0.3, 0.4], "b-")
ax1.scatter([0, 1, 2, 3], [0.5, 0.7, 0.3, 0.4], color="blue")
ax1.text(0.1, 0.6, "Œµ‚ÇÅ")
ax1.text(1.1, 0.8, "Œµ‚ÇÇ")
ax1.text(2.1, 0.4, "Œµ‚ÇÉ")
ax1.text(3.1, 0.5, "Œµ‚ÇÑ")
ax1.set_title("E(œâ) ‚Üí R")
ax1.grid(True)

# 2. –ì—Ä–∞—Ñ–∏–∫ —Å Œµ(1) ‚â• 0.5, Œµ(2) ‚â• 0.2
x_ = np.linspace(0, 4, 100)
ax2.plot(x_, 0.5 + 0.1 * np.sin(x_), "r-", label="Œµ(1) ‚â• 0.5")
ax2.plot(x_, 0.2 + 0.1 * np.cos(x_), "b--", label="Œµ(2) ‚â• 0.2")
ax2.legend()
ax2.grid(True)
ax2.set_title("–ß–∏—Å–ª–æ–≤—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏")

# 3. –ì—Ä–∞—Ñ–∏–∫ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (—Å—Ç—É–ø–µ–Ω—á–∞—Ç—ã–π)
x_steps = np.array([-1, 0, 1, 2, 3, 4])
y_steps = np.array([0, 0, 0.3, 0.6, 0.8, 1])
ax3.step(x_steps, y_steps, "g-", where="post")
ax3.plot(x_steps[1:], y_steps[:-1], "go", fillstyle="none")  # –æ—Ç–∫—Ä—ã—Ç—ã–µ —Ç–æ—á–∫–∏
ax3.plot(x_steps[1:], y_steps[1:], "go")  # –∑–∞–∫—Ä—ã—Ç—ã–µ —Ç–æ—á–∫–∏
ax3.set_title("–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è F(x_)")
ax3.grid(True)

# –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –æ–±—â–µ–≥–æ –≤–∏–¥–∞
plt.tight_layout()
plt.show()
```

–ù–∞ —ç—Ç–∏—Ö –≥—Ä–∞—Ñ–∏–∫–∞—Ö –ø–æ–∫–∞–∑–∞–Ω—ã:

–ü–µ—Ä–≤—ã–π –≥—Ä–∞—Ñ–∏–∫:
–û—Ç–æ–±—Ä–∞–∂–µ–Ω–∏–µ E(œâ) ‚Üí R
–ó–Ω–∞—á–µ–Ω–∏—è Œµ‚ÇÅ, Œµ‚ÇÇ, Œµ‚ÇÉ, Œµ‚ÇÑ

–í—Ç–æ—Ä–æ–π –≥—Ä–∞—Ñ–∏–∫:
–ß–∏—Å–ª–æ–≤—ã–µ —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫–∏
–£—Å–ª–æ–≤–∏—è Œµ(1) ‚â• 0.5 –∏ Œµ(2) ‚â• 0.2

–¢—Ä–µ—Ç–∏–π –≥—Ä–∞—Ñ–∏–∫:
–§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è F(x_)
–°—Ç—É–ø–µ–Ω—á–∞—Ç—ã–π –≤–∏–¥
–û—Ç–∫—Ä—ã—Ç—ã–µ –∏ –∑–∞–∫—Ä—ã—Ç—ã–µ —Ç–æ—á–∫–∏ –≤ –º–µ—Å—Ç–∞—Ö —Å–∫–∞—á–∫–æ–≤

–í—Å–µ –≥—Ä–∞—Ñ–∏–∫–∏ –≤–∫–ª—é—á–∞—é—Ç:
–ö–æ–æ—Ä–¥–∏–Ω–∞—Ç–Ω—ã–µ –æ—Å–∏
–°–µ—Ç–∫—É –¥–ª—è –ª—É—á—à–µ–π —á–∏—Ç–∞–µ–º–æ—Å—Ç–∏
–ü–æ–¥–ø–∏—Å–∏ –∑–Ω–∞—á–µ–Ω–∏–π
–†–∞–∑–ª–∏—á–Ω—ã–µ —Ç–∏–ø—ã –ª–∏–Ω–∏–π –∏ —Ç–æ—á–µ–∫, –∫–∞–∫ –ø–æ–∫–∞–∑–∞–Ω–æ –≤ –∫–æ–Ω—Å–ø–µ–∫—Ç–µ

```python
# –°–æ–∑–¥–∞–µ–º —Ñ–∏–≥—É—Ä—É —Å 4 –ø–æ–¥–≥—Ä–∞—Ñ–∏–∫–∞–º–∏
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 12))

# 1. –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ - PMF
n_, p_ = 20, 0.5  # –ø–∞—Ä–∞–º–µ—Ç—Ä—ã: n_ –∏—Å–ø—ã—Ç–∞–Ω–∏–π, p_ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É—Å–ø–µ—Ö–∞
k_ = np.arange(0, n_ + 1)
binomial = binom.pmf(k_, n_, p_)

ax1.bar(k_, binomial, alpha=0.8, color="blue", label="PMF")
ax1.set_title("–ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ\nProbability Mass Function")
ax1.set_xlabel("–ß–∏—Å–ª–æ —É—Å–ø–µ—Ö–æ–≤ (k_)")
ax1.set_ylabel("–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å")
ax1.grid(True, alpha=0.3)

# 2. –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ - CDF
binomial_cdf = binom.cdf(k_, n_, p_)

ax2.step(k_, binomial_cdf, where="post", color="blue", label="CDF")
ax2.set_title("–ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ\nCumulative Distribution Function")
ax2.set_xlabel("–ß–∏—Å–ª–æ —É—Å–ø–µ—Ö–æ–≤ (k_)")
ax2.set_ylabel("–ù–∞–∫–æ–ø–ª–µ–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å")
ax2.grid(True, alpha=0.3)

# 3. –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ - PDF
x_ = np.linspace(-4, 4, 1000)
mu_values = [-1, 0, 1]  # —Ä–∞–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–≥–æ –æ–∂–∏–¥–∞–Ω–∏—è
sigma_values = [0.5, 1, 1.5]  # —Ä–∞–∑–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–≥–æ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏—è
colors = ["green", "blue", "red"]

for mu, sigma, color in zip(mu_values, sigma_values, colors):
    pdf = norm.pdf(x_, mu, sigma)
    ax3.plot(x_, pdf, color=color, label=f"Œº={mu}, œÉ={sigma}")

ax3.set_title("–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ\nProbability Density Function")
ax3.set_xlabel("x_")
ax3.set_ylabel("–ü–ª–æ—Ç–Ω–æ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏")
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ - CDF
for mu, sigma, color in zip(mu_values, sigma_values, colors):
    cdf = norm.cdf(x_, mu, sigma)
    ax4.plot(x_, cdf, color=color, label=f"Œº={mu}, œÉ={sigma}")

ax4.set_title("–ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ\nCumulative Distribution Function")
ax4.set_xlabel("x_")
ax4.set_ylabel("–ù–∞–∫–æ–ø–ª–µ–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å")
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()  # –í–∞–∂–Ω–æ –¥–æ–±–∞–≤–∏—Ç—å —ç—Ç—É —Å—Ç—Ä–æ–∫—É!
```

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–≥–æ –∏ –Ω–æ—Ä–º–∞–ª—å–Ω–æ–≥–æ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–π

## –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
–ì—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –≤–∫–ª—é—á–∞–µ—Ç:
- PMF (Probability Mass Function - —Ñ—É–Ω–∫—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏)
    - –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ –æ—Ç–¥–µ–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
    - –ü—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∞ —Å—Ç–æ–ª–±—á–∞—Ç–æ–π –¥–∏–∞–≥—Ä–∞–º–º–æ–π
    - –î–∏—Å–∫—Ä–µ—Ç–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è (Œæ‚ÇÅ,Œæ‚ÇÇ,Œæ‚ÇÉ,Œæ‚ÇÑ,Œæ‚ÇÖ)
–ö–∞–∂–¥—ã–π —Å—Ç–æ–ª–±–∏–∫ –ø–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –∫–æ–Ω–∫—Ä–µ—Ç–Ω–æ–≥–æ –∏—Å—Ö–æ–¥–∞
- –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è (CDF - Cumulative Distribution Function)
    - –°—Ç—É–ø–µ–Ω—á–∞—Ç–∞—è —Ñ—É–Ω–∫—Ü–∏—è
    - –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞–∫–æ–ø–ª–µ–Ω–Ω—É—é –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
    - –í—Å–µ–≥–¥–∞ –≤–æ–∑—Ä–∞—Å—Ç–∞–µ—Ç
    - –î–æ—Å—Ç–∏–≥–∞–µ—Ç 1 –≤ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–º –∑–Ω–∞—á–µ–Ω–∏–∏

–§–û–†–ú–£–õ–´:
p_(x_ = k_) = C(n_,k_) * p_^k_ * (1-p_)^(n_-k_)
–≥–¥–µ:
- n_ - —á–∏—Å–ª–æ –∏—Å–ø—ã—Ç–∞–Ω–∏–π
- k_ - —á–∏—Å–ª–æ —É—Å–ø–µ—Ö–æ–≤
- p_ - –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —É—Å–ø–µ—Ö–∞ –≤ –∫–∞–∂–¥–æ–º –∏—Å–ø—ã—Ç–∞–Ω–∏–∏

## –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
–ì—Ä–∞—Ñ–∏—á–µ—Å–∫–æ–µ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –≤–∫–ª—é—á–∞–µ—Ç:
- PDF (Probability Density Function) - —Ñ—É–Ω–∫—Ü–∏—è –ø–ª–æ—Ç–Ω–æ—Å—Ç–∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
    - –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–∞—è –∫–æ–ª–æ–∫–æ–ª–æ–æ–±—Ä–∞–∑–Ω–∞—è –∫—Ä–∏–≤–∞—è
    - –°–∏–º–º–µ—Ç—Ä–∏—á–Ω–∞ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ —Å—Ä–µ–¥–Ω–µ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è (Œº)
    - –†–∞–∑–Ω—ã–µ –∫—Ä–∏–≤—ã–µ –¥–ª—è —Ä–∞–∑–Ω—ã—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ (Œº –∏ œÉ)
    - –ü–æ–∫–∞–∑–∞–Ω—ã –∑–Ω–∞—á–µ–Ω–∏—è 150, 168, 200 (–≤–æ–∑–º–æ–∂–Ω–æ, —Ä–æ—Å—Ç –≤ —Å–º)
- CDF (Cumulative Distribution Function) - —Ñ—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
    - –ü–ª–∞–≤–Ω–∞—è S-–æ–±—Ä–∞–∑–Ω–∞—è –∫—Ä–∏–≤–∞—è
    - –ü–æ–∫–∞–∑—ã–≤–∞–µ—Ç –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –≤–µ–ª–∏—á–∏–Ω–∞ –º–µ–Ω—å—à–µ –∑–∞–¥–∞–Ω–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è
    - –°—Ç—Ä–µ–º–∏—Ç—Å—è –∫ 1 –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ x_

–§–û–†–ú–£–õ–´:
f(x_) = (1/(œÉ‚àö(2œÄ))) * e^(-(x_-Œº)¬≤/(2œÉ¬≤)),  –≥–¥–µ e ‚âà 2.71828... (—á–∏—Å–ª–æ –≠–π–ª–µ—Ä–∞)
–≥–¥–µ:
- Œº (–º—é) - –º–∞—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–æ–µ –æ–∂–∏–¥–∞–Ω–∏–µ
- œÉ (—Å–∏–≥–º–∞) - —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–æ–µ –æ—Ç–∫–ª–æ–Ω–µ–Ω–∏–µ

## –°—Ä–∞–≤–Ω–µ–Ω–∏–µ
### –ù–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ—Å—Ç—å:
- –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ: –¥–∏—Å–∫—Ä–µ—Ç–Ω–æ–µ (—Å—Ç–æ–ª–±–∏–∫–∏)
- –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ: –Ω–µ–ø—Ä–µ—Ä—ã–≤–Ω–æ–µ (–≥–ª–∞–¥–∫–∞—è –∫—Ä–∏–≤–∞—è)
### –§–æ—Ä–º–∞:
- –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ: –º–æ–∂–µ—Ç –±—ã—Ç—å –∞—Å–∏–º–º–µ—Ç—Ä–∏—á–Ω—ã–º
- –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ: –≤—Å–µ–≥–¥–∞ —Å–∏–º–º–µ—Ç—Ä–∏—á–Ω–æ
### –ü—Ä–∏–º–µ–Ω–µ–Ω–∏–µ:
- –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ: –¥–ª—è –ø–æ–¥—Å—á–µ—Ç–∞ —É—Å–ø–µ—Ö–æ–≤ –≤ —Å–µ—Ä–∏–∏ –∏—Å–ø—ã—Ç–∞–Ω–∏–π
- –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ: –¥–ª—è –æ–ø–∏—Å–∞–Ω–∏—è –ø—Ä–∏—Ä–æ–¥–Ω—ã—Ö —è–≤–ª–µ–Ω–∏–π (—Ä–æ—Å—Ç, –≤–µ—Å, —Ç–µ–º–ø–µ—Ä–∞—Ç—É—Ä–∞)

### –§—É–Ω–∫—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è:
- –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ: —Å—Ç—É–ø–µ–Ω—á–∞—Ç–∞—è
- –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ: –≥–ª–∞–¥–∫–∞—è

### –ü—Ä–∞–∫—Ç–∏—á–µ—Å–∫–æ–µ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ:
- –ë–∏–Ω–æ–º–∏–∞–ª—å–Ω–æ–µ: –ø–æ–¥–±—Ä–∞—Å—ã–≤–∞–Ω–∏–µ –º–æ–Ω–µ—Ç—ã, —Ç–µ—Å—Ç—ã —Å –≤—ã–±–æ—Ä–æ–º –æ—Ç–≤–µ—Ç–∞
- –ù–æ—Ä–º–∞–ª—å–Ω–æ–µ: –∏–∑–º–µ—Ä–µ–Ω–∏–µ —Ä–æ—Å—Ç–∞ –ª—é–¥–µ–π, –æ—Ü–µ–Ω–∫–∞ –ø–æ–≥—Ä–µ—à–Ω–æ—Å—Ç–µ–π –∏–∑–º–µ—Ä–µ–Ω–∏–π

# –°–≤–æ–π—Å—Ç–≤–∞ —Ñ—É–Ω–∫—Ü–∏–∏ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è

## –û—Å–Ω–æ–≤–Ω—ã–µ —Å–≤–æ–π—Å—Ç–≤–∞ F(x_)

- –û–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω–æ—Å—Ç—å
0 ‚â§ F(x_) ‚â§ 1
–≠—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ —Å–ª–µ–¥—É–µ—Ç –∏–∑ —Ç–æ–≥–æ, —á—Ç–æ F(x_) –µ—Å—Ç—å –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å

- –ú–æ–Ω–æ—Ç–æ–Ω–Ω–æ—Å—Ç—å
F(x_) - –Ω–µ—É–±—ã–≤–∞—é—â–∞—è —Ñ—É–Ω–∫—Ü–∏—è
–ï—Å–ª–∏ x‚ÇÅ < x‚ÇÇ, —Ç–æ F(x‚ÇÅ) ‚â§ F(x‚ÇÇ)

- –ü—Ä–µ–¥–µ–ª—å–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è
F(-‚àû) = lim[x‚Üí-‚àû] F(x_) = 0
F(+‚àû) = lim[x‚Üí+‚àû] F(x_) = 1

-–í–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª
- p_(Œ± < Œæ < Œ≤) = F(Œ≤) - F(Œ±)
–≠—Ç–æ —Å–≤–æ–π—Å—Ç–≤–æ —Å–≤—è–∑—ã–≤–∞–µ—Ç —Ñ—É–Ω–∫—Ü–∏—é —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è –∏ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–Ω–∏—è –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª
