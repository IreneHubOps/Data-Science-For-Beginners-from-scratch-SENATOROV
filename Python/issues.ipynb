Создание Issues

4. Чтобы создать новое Issue в репозитории на GitHub, следуйте этим шагам:

    1. Перейдите на страницу репозитория, в котором вы хотите создать issue.
    2. Нажмите на вкладку "Issues" (она находится на верхней панели, рядом с "Pull requests").
    3. На странице Issues нажмите на кнопку "New issue" (находится справа).
    4. Введите заголовок (title) для issue, который кратко описывает задачу или проблему.
    5. В поле описания (description) добавьте подробности о проблеме или задаче.
    6. Выберите метки (labels), если необходимо.
    7. Назначьте ответственных (Assignees), если нужно.
    8. При необходимости добавьте связанные проекты или другие атрибуты.
    9. Нажмите "Submit new issue", чтобы создать issue.

5. Подробное описание проблемы: Четко и детально объясните, в чем заключается проблема или что требуется изменить.
Шаги для воспроизведения (для багов): Если это баг, укажите шаги, которые необходимо выполнить, чтобы воспроизвести проблему. Это поможет другим участникам легко понять, как возникла ошибка. Ожидаемое поведение: Укажите, как, по вашему мнению, система должна работать, чтобы проблема была решена или задача была выполнена. Текущие результаты: Опишите, что происходит в текущей реализации. Это поможет отличить ожидаемое поведение от реального. Логи и скриншоты: Приложите логи ошибок, консольные сообщения или скриншоты, которые могут помочь в решении проблемы. Это особенно полезно при решении багов. Информация о системе или среде: В некоторых случаях полезно указать версию программы, операционную систему или другие детали окружения, в которых возникла ошибка или должна быть добавлена новая функциональность.
6. Любые. Тэги могут быть использованы для обозначения типа задачи, приоритета, состояния и т.д.
Стандартные теги:
    1. bug — для обозначения ошибок или багов в коде.
    2. enhancement — для предложений по улучшению существующего функционала.
    3. documentation — для задач, связанных с улучшением документации.
    4. good first issue — для задач, которые подходят для начинающих разработчиков.
    5. help wanted — когда проект нуждается в помощи для решения задачи или проблемы.
    6. question — для обсуждения вопросов или неопределенностей.
    7. wontfix — если задача не будет исправлена или функциональность не будет добавлена.
    8. invalid — если issue не имеет отношения к проекту или неправильное описание.
7. В разделе Assignees можно назначить одного или нескольких участников. Для этого нажмите на "Assign yourself" (назначить себя) или "Assign" (назначить другого пользователя), чтобы выбрать пользователя из списка.

Работа с Issues
8. Как использовать Labels для классификации задач?

8. Labels позволяют удобно фильтровать и искать задачи по различным признакам, таким как тип задачи, приоритет, статус и другие характеристики, соответственно, они должны быть максимально по смыслу приближены к issues. Можно использовать стандартные тэги, либо написать свои
9. Отслеживание прогресса: Milestones помогают организовать задачи по этапам разработки. Например, можно создать milestone для релиза новой версии программы. Ожидаемые даты: Для каждого milestone можно установить предполагаемую дату завершения, что помогает отслеживать сроки или понимать, в какие сроки задача должна быть решена. Справа на странице issue, в разделе "Milestones", выберите нужный milestone из списка.
Если вы еще не создали milestone, вы можете создать новый, нажав на "Create a new milestone" и указав название и дату завершения.
10. В описании Pull Request или сообщении коммита используйте ключевые слова и номер Issue, например:
    Fixes #123
    Closes #123
    Resolves #123
Эти ключевые слова указывают GitHub, что данный Pull Request решает или закрывает Issue с номером 123. Когда Pull Request будет принят (мержится в основную ветку), связанное Issue будет автоматически закрыто.
11. Перейдите на страницу Issue, к которому хотите добавить комментарий. Внизу страницы, под списком комментариев, вы увидите поле для ввода текста "Write a comment". Напишите свой комментарий и нажмите кнопку "Comment".

18. Настройки уведомлений в GitHub:
Перейдите в Settings вашего аккаунта на GitHub. В разделе Notifications выберите, как именно вы хотите получать уведомления: по электронной почте, через веб-уведомления или через мобильные устройства.
Вы можете подписаться на конкретные Issues или репозитории. Перейдите к Issue, на которое хотите подписаться. Нажмите на "Subscribe" в правом верхнем углу страницы Issue. Это позволит получать уведомления о новых комментариях и изменениях в этом Issue.
В Settings каждого репозитория можно настроить, какие события будут отслеживаться и какие уведомления будут отправляться. Например, можно настроить уведомления для всех новых Issues или при изменениях статуса.
19.  Проект представляет собой доску (board) с колонками, в которых можно разместить задачи и отслеживать их выполнение. Это помогает организовать работу и связать Issues с конкретными задачами в проекте. В меню "Projects" выберите "New Project". Дайте проекту название и выберите тип. В открытом проекте нажмите на "Add cards". В появившемся списке выберите Issues или Pull Requests, которые хотите добавить в проект. Для каждой карточки (задачи) можно указать статус, срок выполнения и т. д. Проекты GitHub позволяют автоматически перемещать Issues между колонками в зависимости от их статуса. Например, можно настроить автоматические переходы из "To Do" в "In Progress" или "Done".
20. Probot — это популярная платформа для создания ботов на GitHub. С помощью Probot можно настроить автоматические действия с Issues, например: приветствие новых участников репозитория, автоматическое присваивание меток или исполнителей на основе событий в Issues и т.д.

Webhooks позволяют интегрировать GitHub с внешними сервисами и автоматически выполнять действия, основываясь на событиях в репозиториях. ZenHub: интегрируется с GitHub и добавляет функциональность для управления проектами, планирования спринтов, визуализации задач и более удобного взаимодействия с Issues. Практические аспекты
24. Как использовать шаблоны для создания Issues?

24. Шаблоны для Issues на GitHub помогают создать стандартизированные формы для различных типов задач или ошибок. Это полезно для поддержания порядка и улучшения процесса создания Issues. Необходимо строго следовать указаниям и заполнять поля в соответствии с ними.
25. Linked Issues — это функция GitHub, которая позволяет создавать связи между различными Issues. Это полезно для отслеживания взаимозависимостей между задачами, например, когда одно Issue зависит от выполнения другого. 
В комментариях к Issue используйте ключевые слова, такие как:
    1. Closes #номер_проблемы — автоматически закроет связанное Issue при закрытии текущего.
    2. Fixes #номер_проблемы — автоматически закроет связанное Issue при закрытии текущего.
    3. Related to #номер_проблемы — создает связь между задачами без автоматического закрытия.
    4. Depends on #номер_проблемы — указывает, что задача зависит от выполнения другого Issue.
26. Время выполнения: Хотя GitHub не предоставляет встроенной функции для отслеживания времени, многие пользователи добавляют это вручную, используя комментарии или специальные метки (labels). С помощью меток можно классифицировать Issues по типам, сложности или состоянию. Это помогает отслеживать прогресс и приоритет задач. Milestones позволяют группировать Issues по версии или спринту, что помогает отслеживать, сколько задач выполнено за определённый период.Использование GitHub Projects помогает визуализировать работу, отслеживая задачи и их прогресс по колонкам (например, "To Do", "In Progress", "Done"). Использование GitHub Actions для автоматизации может помочь отслеживать метрики, такие как время на выполнение CI/CD пайплайнов, количество выполненных тестов и другие параметры.
27. Best practices включают четкое описание задач, использование меток и назначение исполнителей, а также регулярную проверку и автоматизацию процессов.
