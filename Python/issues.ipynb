#### Общие вопросы

    1. Что такое Issues на GitHub и для чего они используются?

 Это инструмент для работы с репозиториями в GitHib, который помогает
 планировать, обсуждать и отслеживать работу, такую как устранение багов,
 добавление фич, улучшение качества кода. 

    2. Чем Issues отличаются от других инструментов управления задачами?

Специализирован на работе с объектами GitHub (например, репо, ветки, коммиты,
пулл-реквесты, другие ишьюсы) и делает удобным доработку проектов и дальнейшее
отслеживание изменений

    3. Какие основные компоненты (поля) есть у каждого Issue?

Title - заголовок, checklist - чек-лист создания issue, body - описание,
milestone - дедлайн или спринт, assignees - исполнители, labels - теги,
type - тип (баг, фича или задача), project - связи с проектами

#### Создание Issues

    4. Как создать новое Issue в репозитории?

Заходим в репозиторий - раздел issues - нажимаем New issue

    5. Какие данные рекомендуется указывать в описании Issue для лучшего
понимания задачи?

Заголовок, описание проблемы детальное со вставкой проблемного кода или ссылки
на фрагмент, с которым связан ишью, при отсутствии проекта в GitHub
необходимо добавить zip-файл с кодом, визуальные демо ошибок (видео, фото),
другая необходимая инфо для понимания ишью или поиска проблемы (например,
описание шагов)

    6. Какие теги (labels) можно добавить к Issue? Какие из них стандартные?

bug (стандарт), documentation (стандарт), duplicate, enhancement (стандарт),
good first issue, help wanted, invalid, question (стандарт), wontfix
Также можно делать кастомные теги

    7. Как прикрепить Assignees (ответственных) к Issue?

Нажать на шестеренку в правом сайдбаре, кликнуть на нужного пользователя

#### Работа с Issues

    8. Как использовать Labels для классификации задач?

Выбрать соответствующий тег в правом сайдбаре или создать свой тег,
отражающий особенности issue, которые нельзя отнести к стандартным тегам

    9. Для чего нужен Milestone, и как связать его с Issue?

Milestone нужен для помощи в приоритизации выполнения issue, выбрать в правом
сайдбаре с имеющимся milestone (или создать свой milestone, или просто
добавить дату дедлайна)

    10. Как привязать Issue к пул-реквесту (Pull Request)?

Вставляем в описание пулл-реквеста #номер_issue с ключевыми словами. Например:

    close
    закрывается
    закрыто
    fx
    ошибок
    fixed
    resolve
    разрешается
    разрешенные

    11. Как добавить комментарий к существующему Issue?

Внизу issue - add a comment - comment

#### Закрытие и завершение Issues

    12. Как закрыть Issue вручную?

Внизу issue под комментарием выбрать close issue 

    13. Можно ли автоматически закрыть Issue с помощью сообщения в коммите или
пул-реквесте? Как это сделать?

С помощью ключевого слова close #номер_issue в описании пул-реквеста

    14. Как повторно открыть закрытое Issue, если работа ещё не завершена?

Внизу issue под комментарием выбрать reopen issue

#### Фильтрация и поиск

    15. Как найти все открытые или закрытые Issues в репозитории?

Раздел Issues в репозитории - использование фильтров open/closed

    16. Как использовать фильтры для поиска Issues по меткам, исполнителям
или другим критериям?

Раздел Issues в репозитории - использование фильтров labels, assignees и др.

    17. Как сортировать Issues по приоритету, дате создания или другим
параметрам?

Раздел Issues в репозитории - справа кнопка Sort, где можно выбрать вид
сортировки

#### Интеграции и автоматизация

    18. Как настроить автоматические уведомления о новых или изменённых
Issues?

Для issue - в правом сайдбаре notifications - subscribe
Для всего репо можно выбрать watch

    19. Что такое Projects в контексте GitHub, и как связать их с Issues?

Projects - инструмент для планирования и отслеживания работы в репозитории
В правом сайдбаре выбрать нужный project (или создать его предварительно)

    20. Какие сторонние инструменты можно использовать для автоматизации
работы с Issues (например, боты, Webhooks)?

Workflow в Actions, которые как правило направлены на работу с деплоем,
безопасностью, непрерывной интеграциией, автоматизацией, страницами

#### Коллаборация

    21. Как упомянуть другого пользователя в комментарии к Issue?

С помощью @

    22. Как запросить дополнительные данные или уточнения у автора Issue?

Создать комментарий с упоминанием автора issue, если автор подписан на
issue, то увидит запрос самостоятельно, даже если его не упомянули

    23. Что делать, если Issue неактуально или его нужно объединить с другим?

Внизу issue выбрать close as duplicate, также можно упомянуть #номер_issue
в описании

#### Практические аспекты

    24. Как использовать шаблоны для создания Issues?

Добавить шаблоны в папку .github - ISSUE_TEMPLATE - при создании ишью выбрать
шаблон или сделать новый

    25. Что такое Linked Issues, и как создать связь между задачами?

Связанный ишью, который дополняет ишью, дублирует или отменяет. Для связи
необходимо прописать #номер_issue, если они находятся с одном репо или же
использовать ссылку на связанный ишью в описании ишью. Или же добавить
ишью в правом сайдбаре relationships

    26. Какие метрики (например, время выполнения) можно отслеживать с помощью
Issues?

Количество открытых и закрытых задач, распределение issue по исполнителям,
типам, проектам, авторам и milestones, время выполнения задач
Также можно использовать чек-лист, где можно отслеживать статусы выполнения
issue
Можно настроить Insights и Actions для отслеживания issues

    27. Какие best practices рекомендуются при работе с Issues в команде?

Точное название ишью и описание, включение кода с помощью ``, добавление
визуализации (видео, скрины), ссылки на код, к которому обращается issue,
использование ключевых слов, использование тегов открытия/закрытия issue в ТГ
